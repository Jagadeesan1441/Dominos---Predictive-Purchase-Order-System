```python
def outliers_removal(newmerged_df):
  for i in newmerged_df[["unit_price", "total_price", "quantity"]]:
    q1=newmerged_df[i].quantile(0.15)
    q3=newmerged_df[i].quantile(0.85)
    iqr=q3-q1
    lower_limit=q1-1.5*iqr
    upper_limit=q3+1.5*iqr
    newmerged_df=newmerged_df[(newmerged_df[i]>=lower_limit) & (newmerged_df[i]<=upper_limit)]
  return newmerged_df
```

```python

from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# Prepare daily sales data
daily_sales = df.groupby(df['order_date'].dt.date)['total_price'].sum()
daily_sales.index = pd.to_datetime(daily_sales.index)
daily_sales = daily_sales.resample('D').sum().fillna(0).asfreq('D')

# Apply seasonal decomposition
result = seasonal_decompose(daily_sales, model='additive')

# Plot the results
fig, axes = plt.subplots(3, 1, figsize=(14, 10))
titles = ['Observed', 'Trend', 'Seasonal']
for i, component in enumerate([result.observed, result.trend, result.seasonal]):
    component.plot(ax=axes[i], title=titles[i], xlabel='Date', ylabel='Sales', fontsize=14)
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
def prepare_weekly_sales(df):
  if "order_date" not in pizza.columns:
    raise ValueError("The DataFrame must have a 'order_date' column.")
  if "pizza_name_id" not in pizza.columns:
    raise ValueError("The DataFrame must have a 'pizza_name_id' column.")

  df["order_date"]=pd.to_datetime(df["order_date"])
  df["week"]=df["order_date"].dt.to_period("W").apply(lambda r:r.start_time)
  weeklysales=df.groupby(["week","pizza_name_id"])["quantity"].sum().reset_index()
  weeklysales.rename(columns={"week":"ds","quantity":"y"},inplace=True)
  return weeklysales
```

```python
dominos=prepare_weekly_sales(pizza)
dominos.head()
```

```python
from sklearn.metrics import mean_absolute_percentage_error as mape


def train_models_for_all_pizza(dominos):
  pizza_models={}
  pizza_forecast={}
  pizza_mape_scores=[]

  for pizza_id in dominos["pizza_name_id"].unique():
    pizza_df=dominos[dominos["pizza_name_id"]==pizza_id]

    train_size=int(len(pizza_df)*0.8)
    train,test=pizza_df[:train_size],pizza_df[train_size:]


    model=Prophet()
    model.fit(train)

    future=model.make_future_dataframe(periods=len(test),freq="W")
    forecast=model.predict(future)


    prediction=forecast["yhat"][-len(test):].values
    prophet_mape=mape(test["y"].values,prediction)
    pizza_mape_scores.append({"pizza_name_id":pizza_id,"mape":prophet_mape})

    pizza_models[pizza_id]=model
    pizza_forecast[pizza_id]=forecast[["ds","yhat"]].tail(len(test)).reset_index(drop=True)


  all_forecast=pd.concat([pd.DataFrame({"pizza_name_id":pizza_id,"ds":pizza_forecast[pizza_id]["ds"],"yhat":pizza_forecast[pizza_id]["yhat"]})  for pizza_id in pizza_forecast],ignore_index=True)

  return pizza_models,pizza_mape_scores,all_forecast
```

```python
pizza_models,pizza_mape_scores,all_forecast=train_models_for_all_pizza(dominos)
```

```python
import pickle

# assuming your model is named `model`
with open('pizza_model.pkl', 'wb') as file:
    pickle.dump(pizza_models, file)
print("Model saved as 'pizza_model.pkl'")
```

#INGREDIENTS AND PURCHASE ORDER CREATION

```python
import pandas as pd
```

```python
ingridients_per_pizza=pd.read_csv('/content/dominos_pizzadata')
```

```python
ingridients_per_pizza.columns
```

```python
ingredients_demand=pd.merge(all_forecast,ingridients_per_pizza,on="pizza_name_id",how="inner")
```

```python
ingredients_demand["total_quantity_required"]=ingredients_demand["yhat"]*ingredients_demand["quantity"]
ingredients_demand.head(5)
```

```python
total_ingridient_demand=ingredients_demand.groupby("pizza_name_id")["total_quantity_required"].sum().reset_index()
```

```python
total_ingridient_demand=total_ingridient_demand.sort_values(by="total_quantity_required",ascending=False)
total_ingridient_demand["total_quantity_reqkg"]=total_ingridient_demand["total_quantity_required"]/1000
```

```python
purchase_order=pd.merge(total_ingridient_demand,ingridients_per_pizza[["pizza_name_id","pizza_ingredients","pizza_name","Items_Qty_In_Grams"]].drop_duplicates(),on="pizza_name_id",how="left")
```

```python
purchase_order=purchase_order.rename(columns={"pizza_name_id":"pizzatype","total_quantity_reqkg":"Ingridients (kg)"})
```

```python
purchase_order=purchase_order.groupby(["pizzatype","pizza_name","pizza_ingredients"])["Ingridients (kg)"].sum().reset_index()
```
